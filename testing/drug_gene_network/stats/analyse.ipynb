```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
```

```python
jacc = pd.read_csv('jaccard_indices.csv')
cos = pd.read_csv('embedding_similarity.csv')
stats = jacc.merge(cos, on=['drug1', 'drug2'])
stats
```

# Jaccard index of the weighted projection network

```python
x = 'weighted_jaccard'
y = 'cosine_sim'
corr = pearsonr(stats[x], stats[y])
sns.scatterplot(data=stats, x=x, y=y)
plt.title(f'{y.replace("_", " ")} vs {x.replace("_", " ")}. Corr = {round(corr[0], 3)} ')
plt.show()
```
